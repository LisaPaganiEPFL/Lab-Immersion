### Plot the mean of the dynamics

```python
def extract_ratio_fix(path,x,n):
    
    deme_dynamics = []
    
    for i in range(n):
        path_sub = path + '/fix_' + str(i+1) + '.npz'
        fix = np.load(path_sub)['arr_0']
        sum_tot = np.sum(fix[:x], axis=1)
        deme_dynamics.append(sum_tot.T[0]/(sum_tot.T[0]+sum_tot.T[1]))
    
    return np.mean(deme_dynamics,axis=0)

def extract_ratio_ex(path,x,n):
    
    deme_dynamics = []
    
    for i in range(n):
        path_sub = path + '/ex_' + str(i) + '.npz'
        fix = np.load(path_sub)['arr_0']
        sum_tot = np.sum(fix[:x], axis=1)
        deme_dynamics.append(sum_tot.T[0]/(sum_tot.T[0]+sum_tot.T[1]))
    
    return np.mean(deme_dynamics,axis=0)
```

```python
x=2500
star_fix_leaf = extract_ratio_fix("Test5/Leaf/Star",x,999)
star_ex_leaf = extract_ratio_ex("Test5/Leaf/Star",x,1000)
clique_fix_leaf = extract_ratio_fix("Test5/Leaf/Clique",x,999)
clique_ex_leaf = extract_ratio_ex("Test5/Leaf/Clique",x,1000)

star_fix_center = extract_ratio_fix("Test5/Center/Star",x,636)
star_ex_center = extract_ratio_ex("Test5/Center/Star",x,1000)
clique_fix_center = extract_ratio_fix("Test5/Center/Clique",x,999)
clique_ex_center = extract_ratio_ex("Test5/Center/Clique",x,1000)

wm_fix = extract_ratio_fix("Test5/Wm",x,999)
wm_ex = extract_ratio_ex("Test5/Wm",x,1000)
```

```python
center = 'magenta'
leaf1 = 'chartreuse'
leaf2 = 'royalblue'

plt.plot(np.arange(1,x+1,1),star_fix_leaf,color=center,label='Star')
plt.plot(np.arange(1,x+1,1),clique_fix_leaf,color=leaf1,label='Clique')
plt.plot(np.arange(1,x+1,1),wm_fix,color=leaf2,label='Wm')
plt.title('Fixation mean trajectories (leaf)')
plt.legend()
plt.show()

plt.plot(np.arange(1,x+1,1),star_fix_center,color=center,label='Star')
plt.plot(np.arange(1,x+1,1),clique_fix_center,color=leaf1,label='Clique')
plt.plot(np.arange(1,x+1,1),wm_fix,color=leaf2,label='Wm')
plt.title('Fixation mean trajectories (center)')
plt.legend()
plt.show()
```

```python
def weight(center, leaf):
    
    K = 1000
    N_demes = 5
    Nc = ((N_demes-1)*mI + (1-(N_demes-1)*mO))*K
    Nl = (mO + 1-mI)*K

    return (center * Nc + leaf * (N_demes-1)*Nl)/(K*N_demes)
```

```python
def extract_ratio_fix(path,x,n):
    
    deme_dynamics = []
    
    for i in range(n):
        path_sub = path + '/fix_' + str(i+1) + '.npz'
        fix = np.load(path_sub)['arr_0']
        sum_tot = np.sum(fix[:x], axis=1)
        deme_dynamics.append(sum_tot.T[0]/(sum_tot.T[0]+sum_tot.T[1]))
    
    return deme_dynamics

def extract_ratio_ex(path,x,n):
    
    deme_dynamics = []
    
    for i in range(n):
        path_sub = path + '/ex_' + str(i) + '.npz'
        fix = np.load(path_sub)['arr_0']
        sum_tot = np.sum(fix[:x], axis=1)
        deme_dynamics.append(sum_tot.T[0]/(sum_tot.T[0]+sum_tot.T[1]))
    
    return deme_dynamics
```
